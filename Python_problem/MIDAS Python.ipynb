### Authentication

```python
auth = OAuth1(api_key, api_secret_key, access_token, access_token_secret)
url = 'https://api.twitter.com/1.1/account/verify_credentials.json'
requests.get(url, auth=auth)
```

### Get URL of the `user`  

```python
def getURL(screen_name):
    URL = 'https://api.twitter.com/1.1/statuses/user_timeline.json?screen_name={}'.format(screen_name)
    return URL
```

### Number of Tweets from the `user`

```python
username = "midasIIITD"
URL = getURL(username)
tweets_request = requests.get(URL, auth=auth)
tweet_count = tweets_request.json()[0]["user"]["statuses_count"]
```

### Get Tweets

```python
URL = URL + '&tweet_mode=extended&count={}'.format(tweet_count)
tweets_request = requests.get(URL, auth=auth)
tweets = tweets_request.json()
```

### Count the `images` in a tweet (including re-tweets)

```python
def images_count(tweet):
    count = 0
    if 'media' in tweet['entities']:
        count = len(tweet['entities']['media'])
    if 'extended_entities' in tweet:
        if 'media' in tweet['extended_entities']:
            count = count + len(tweet['entities']['media'])
    return count
```

### Representation of Data

```python
col_labels = ["Date/Time","Text","Favorite","Retweets","Image_Count"]
```

```python
pd.set_option('display.max_colwidth', -1)
data_tabular = pd.DataFrame(columns = col_labels)
for i, tweet in enumerate(tweets):
    
    image_count = images_count(tweet)
    if image_count == 0:
        count = 'None'
    else:
        count = str(image_count)
    
    data_tabular.loc[i] = [
        tweet['created_at'],
        tweet['full_text'],
        tweet['favorite_count'],
        tweet['retweet_count'],
        count
                          ]

    
data_tabular
```
